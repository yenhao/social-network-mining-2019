Here, we will use Facebook combined ego networks dataset, it contains the aggregated network of ten individuals’ Facebook friends list. You can download the required `facebook_combined.txt` file from the [Stanford University site](https://snap.stanford.edu/data/egonets-Facebook.html).

```python
import pandas as pd
```

```python
# fb_df = nx.read_edgelist("facebook_combined.txt", create_using = nx.Graph(), nodetype=int)

fb_df = pd.read_csv('data/facebook_combined.txt', header=None, names=['node1', 'node2'], sep=' ')
fb_df.head()
```

```python
plt.figure(figsize=(20,20))
nx.draw_networkx(fb_graph)
plt.show()
```

With different [layout](https://networkx.github.io/documentation/stable/reference/drawing.html)

```python
pos = nx.spring_layout(fb_graph)
betCent = nx.betweenness_centrality(fb_graph, normalized=True, endpoints=True)
node_color = [20000.0 * fb_graph.degree(v) for v in fb_graph]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(fb_graph, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off')
```

```python
pr = nx.pagerank_numpy(fb_graph, alpha=0.9)
for k,v in sorted(pr.items(), key=lambda x: x[1], reverse=True)[:100]:
    print(k, v, sep='\t')
```

## Graph Convolutional Networks

```python
# get node order from karate club
order = sorted(list(G.nodes()))
```

```python
# get adjacency matrix 
A = nx.to_numpy_matrix(G, nodelist=order)

# self direction 
I = np.eye(G.number_of_nodes()) 

# final adj
A_hat = A + I 
```

## normalization

```python
D_hat = np.array(np.sum(A_hat, 1)).reshape(-1)  
D_hat = np.matrix(np.diag(D_hat))
```

### Weights for Networks

```python
W_1 = np.random.normal(loc=0, scale=1, size=(G.number_of_nodes(), 4)) # (nodes, hidden)
W_2 = np.random.normal(loc=0, size=(W_1.shape[1], 2)) # (hidden, out_dim)
```

### Final Networks

```python
def gcn_layer(A_hat, D_hat, X, W):
    return relu(D_hat**-1 * A_hat * X * W)

def relu(x): # 定義 ReLU 函數
            return np.maximum(x,0)
```

### Run Network

```python
H_1 = gcn_layer(A_hat, D_hat, I, W_1)
H_2 = gcn_layer(A_hat, D_hat, H_1, W_2)
output = H_2

output
```

```python
labels = [v['club'] for k, v in sorted(list(G.nodes(data=True)))]

label_map = {k:i for i, k in enumerate(set(labels))}
label_map
```

```python
label_idx = [label_map.get(l) for l in labels]
```

```python
output = np.asarray(output)
```

```python
plt.scatter(output[:, 0], output[:, 1], c=label_idx)
plt.show()
```
